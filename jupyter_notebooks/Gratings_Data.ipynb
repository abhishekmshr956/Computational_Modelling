```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
```

```python
Luk_data = np.load('data/Lukas_Data.npy')
#Luk_data.shape 
```

### Lukas Data
Shape = (3, 5, 2, 4000).
1st dim: Stimulus (0=Grating, 1=ON Edge, 2=OFF Edge)
2nd dim: Velocity: 15,30,60,120,240 deg/sec.
3rd dim: PD, ND
4th dim: time in msec

```python
gratings_voltage_data = Luk_data[0, :, :]
```

```python
f, ax = plt.subplots(1,5, sharey=True, figsize=(15,5))
[ax[i].plot(gratings_voltage_data[i][0]) for i in range(5)];
[ax[i].plot(gratings_voltage_data[i][1], color='red') for i in range(5)];
```

### Bulle ArcLightData 

```python
Bulle_arclight_grating_data = pd.read_csv("data/Bulle/ArcLight_Grating_TimeSeries.txt",sep="\t")
column = ['time']
velocity = [15, 30, 60, 120]
[column.extend(['Mean_PD_'+str(velocity[i]),'SEM_PD_'+str(velocity[i])]) for i in range(len(velocity))];
[column.extend(['Mean_ND_'+str(velocity[i]),'SEM_ND_'+str(velocity[i])]) for i in range(len(velocity))];
Bulle_arclight_grating_data.columns = column
#Bulle_arclight_grating_data
```

```python
f, ax = plt.subplots(1,4, sharey=True, figsize=(15,5))
[ax[i].plot(Bulle_arclight_grating_data['Mean_PD_'+str(velocity[i])]) for i in range(4)];
[ax[i].plot(Bulle_arclight_grating_data['Mean_ND_'+str(velocity[i])], color='red') for i in range(4)];
```
