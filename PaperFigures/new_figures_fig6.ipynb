```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['axes.facecolor'] = '#EEEEEE'
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['lines.linewidth'] = 2.0
#plt.style.use('ggplot')

plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)
```

```python
T4_param_label=['trld [%]','HP tc [s]','LP'+r'$_1$'+' tc [s]','LP'+r'$_2$'+' tc [s]','gain [*100]']
T4_param = np.zeros((2,10))
T4_param[0] = [-0.365, 0.015, 0.810, 0.029, 2.715, 0.328, 3.124, 0.897, 189.1, 7.9] #data, standard deviation
#T4_param[0] = [-0.365, 0.005, 0.810, 0.010, 2.715, 0.109, 3.124, 0.299, 189.1, 2.6] #data, sem
T4_param[1] = [-0.215, 0.007, 0.242, 0.020, 1.367, 0.078, 1.370, 0.078, 135.3, 44.685] 
#T4_param[1] = [-0.215, 0.007, 0.242, 0.020, 1.367, 0.078, 1.370, 0.078, 135.3, 44.685]
```

```python
def setmyaxes(myxpos, myypos, myxsize, myysize):
    ax = plt.axes([myxpos, myypos, myxsize, myysize])
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
```

```python
def plot_subfig(data, label, cell_label, cell_color):
    
    mywidth = 0.36
    mydim=5
    
    param_data = data[0:10:2]
    param_data[4] = param_data[4]/100.0
    param_error = data[1:10:2] # / np.sqrt(9) #division if plot sem
    param_error[4] = param_error[4]/100.0
    
    plt.bar(np.arange(mydim),param_data,yerr=param_error,width=mywidth,color=cell_color,label=cell_label)
    plt.xticks(np.arange(mydim),label)
    plt.plot([-0.5,mydim-0.5],[0,0],linestyle='dashed',color='black')
    plt.ylabel('parameter value')
    plt.legend(loc=2,frameon=False,fontsize=14)
    plt.ylim(-0.5,4.5)
    #print(param_data, param_error)
    
def plot_fig6():
    f = plt.figure(figsize=(12,6))
    xsize=0.34
    ysize=0.6
    xoffs=0.40
    yoffs=0.15
    leftm=0.08
    cell_label=['GCaMP6f','GCaMP8f']
    cell_color=['red','blue']
    
    for i in range(2):
        #plt.subplot(1,2,i+1)
        setmyaxes(i*xoffs+leftm, yoffs,xsize,ysize)
        plot_subfig(T4_param[i].copy(),T4_param_label, cell_label[i],cell_color[i])
        
    f.text(0.06,0.77,'A',weight='bold',fontsize=12)
    f.text(0.46,0.77,'B',weight='bold',fontsize=12)
        
    plt.savefig('figures/new_figures/figure6.pdf',dpi=1000, bbox_inches='tight')
plot_fig6()
```

```python
label=['T4c PD','Mi1']
model_misfit = np.zeros((2,2))
model_misfit[0] = [8.465, 4.616] #model misfit for Lmodel T4c, Mi1
model_misfit[1] = [5.771, 6.819] #model misfit for Mmodel T4c, Mi1
```

```python
def plot_subfig9(data, label, cell_label, cell_color):
    
    mywidth = 0.4
    mydim=2
    
    plt.bar(np.arange(mydim),data[0],width=mywidth,color=cell_color[0],label=cell_label[0])
    plt.bar(np.arange(mydim)+mywidth,data[1],width=mywidth,color=cell_color[1],label=cell_label[1])
    plt.xticks(np.arange(mydim)+mywidth/2.0,label)
    #plt.plot([-0.5,mydim-0.5],[0,0],linestyle='dashed',color='black')
    plt.ylabel('model error [% of data power]')
    plt.legend(loc=0,frameon=False,fontsize=14)
    plt.ylim(0.0,9.0)
    #print(param_data, param_error)
    
def plot_fig9():
    plt.figure(figsize=(6,6))
    
    cell_label=['Lmodel','Mmodel']
    cell_color=['blue','green']
    
    plt.subplot(1,1,1)
    plot_subfig9(model_misfit.copy(),label, cell_label,cell_color)
        
    plt.savefig('figures/new_figures/figure9.pdf',dpi=1000)
```

```python
plot_fig9()
```
